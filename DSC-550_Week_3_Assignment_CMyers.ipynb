```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Display three charts using 3 different charts with complete and professional labeling

#Data from matplotlib tutorial
data = {'apple': 10, 
        'orange': 15, 
        'lemon': 5, 
        'lime': 20}

#Create two separate lists for the key value pairs from the dictionary that houses the data.
names = list(data.keys())
values = list(data.values())

#Instantiate figure sizes
plt.figure(figsize=(15,5))

#Create first plot as a bar chart
plt.subplot(131)
plt.bar(names, values)
plt.title('Bar chart of fruits')
plt.xlabel('Fruits')
plt.ylabel('Number of fruits')

#Create second plot as a scatter plot
plt.subplot(132)
plt.scatter(names, values)
plt.title('Scatter plot of fruits')
plt.xlabel('Fruits')
plt.ylabel('Number of fruits')

#Create third plot as a line graph
plt.subplot(133)
plt.plot(names, values,'r--', linewidth=3.0)
plt.title('Line graph of fruits')
plt.xlabel('Fruits')
plt.ylabel('Number of fruits')

#Give the three plots an overarching title
plt.suptitle('Graphs of fruits')

#Adjust the width in between the plots
plt.subplots_adjust(wspace=.25)
```
