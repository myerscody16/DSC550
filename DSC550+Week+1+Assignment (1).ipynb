```python
import pandas as pd
import numpy as np
```

```python
array = np.random.randint(1, 500,(10, 50))
print(array)
```

```python
df = pd.DataFrame(data = array)
print(df)
```

```python
#Added a column where each element is the sum of that row
sumOfRows = df.sum(axis=0)
print(sumOfRows)
```

```python
sumOfColumns = df.sum(axis = 1)
print(sumOfColumns)
```

```python
sumOfAllEntries = df.values.sum()
s = df.sum().sum()
print(s)
print(sumOfAllEntries)
```

```python
#Find max value in dataframe
allMaxValues = df.max()
listOfMaxValues = []
for i in allMaxValues:
    listOfMaxValues.append(i)
print(max(listOfMaxValues))
print(df.max().max())
```

```python
#Find min value in dataframe
allMinValues = df.min()
listOfMinValues = []
for i in allMinValues:
    listOfMinValues.append(i)
print(min(listOfMinValues))
print(df.min().min())
```

```python
#Find average value of all entries
print(df.mean().mean())
```

```python
#Find standard deviation for all entries
print(df.std().std())
```

```python
#sort by rows
df.sort_values(by = [0, 1, 2], axis=0)
```

```python
#sort by columns
df.sort_values(by = [0, 1, 2], axis=1)
```

```python
#sort by row 2
df.sort_values(by=[1], axis = 0)
```

```python
#sort by column 2
df.sort_values(by=[1], axis = 1)
```
